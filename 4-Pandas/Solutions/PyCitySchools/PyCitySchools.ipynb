```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
school_data.rename(columns={"School ID": "school_id"}, inplace=True)

student_data = pd.read_csv(student_data_to_load)
student_data.rename(columns={"Student ID": "student_id"}, inplace=True)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the total number of schools

* Calculate the total number of students

```python
tot_schools = len(school_data)
tot_students = len(student_data)
tot_budget = school_data.budget.sum()
avg_reading = student_data.reading_score.mean()
avg_math = student_data.reading_score.mean()
```

```python
mask = student_data.math_score >= 70
math_percent = (len(student_data.loc[mask]) / tot_students) *100
```

```python
mask = student_data.reading_score >= 70
reading_percent = (len(student_data.loc[mask]) / tot_students) *100
```

```python
mask = (student_data.math_score >= 70) & (student_data.reading_score >= 70)
both_percent = (len(student_data.loc[mask]) / tot_students) *100
```

```python
output = {
    "Total Schools": tot_schools,
    "Total Students": tot_students,
    "Total Budget": tot_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Math": math_percent,
    "% Passing Reading": reading_percent,
    "Overall Passing Rate": both_percent
}
summarytable = pd.DataFrame([output])
summarytable
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
complete_copy = school_data_complete.copy()
school_copy = school_data.copy()
```

```python
tot_students = complete_copy.groupby(["school_name", "type"]).student_id.count()
tot_budget = school_copy.groupby(["school_name", "type"]).budget.sum()
per_budget = tot_budget / tot_students
```

```python
avg_math = complete_copy.groupby(["school_name", "type"]).math_score.mean()
avg_reading = complete_copy.groupby(["school_name", "type"]).reading_score.mean()
per_math = (complete_copy.loc[complete_copy.math_score >= 70].groupby(["school_name", "type"]).student_id.count()) / tot_students
per_reading = (complete_copy.loc[complete_copy.reading_score >= 70].groupby(["school_name", "type"]).student_id.count()) / tot_students
per_both = (complete_copy.loc[(complete_copy.math_score >= 70) & (complete_copy.reading_score >= 70)].groupby(["school_name", "type"]).student_id.count()) / tot_students
per_math = per_math *100
per_reading = per_reading *100
per_both = per_both *100
```

```python
summarytable2 = pd.DataFrame()
summarytable2["Total Students"] = tot_students
summarytable2["Total Budget"] = tot_budget
summarytable2["Per Student Budget"] = per_budget
summarytable2["Average Math"] = avg_math
summarytable2["Average Reading"] = avg_reading
summarytable2["% Passing Math"] = per_math
summarytable2["% Passing Reading"] = per_reading
summarytable2["% Passing Both"] = per_both
summarytable2
```

```python
nine_math = student_data.loc[student_data.grade == "9th"].groupby("school_name").math_score.mean()
ten_math = student_data.loc[student_data.grade == "10th"].groupby("school_name").math_score.mean()
eleven_math = student_data.loc[student_data.grade == "11th"].groupby("school_name").math_score.mean()
twelve_math= student_data.loc[student_data.grade == "12th"].groupby("school_name").math_score.mean()
```

```python
summarytable3 = pd.DataFrame()
summarytable3["Average 9th Math Scores"] = nine_math
summarytable3["Average 10th Math Scores"] = ten_math
summarytable3["Average 11th Math Scores"] = eleven_math
summarytable3["Average 12th Math Scores"] = twelve_math
summarytable3
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 630, 645, 680]
bin_name = ["<$585", "$585-$630", "$630-$645", "$645-$680"]
summarytable5 = summarytable2.copy()
summarytable5["Spending Ranges (Per Student)"] = pd.cut(summarytable5["Per Student Budget"], bins, labels = bin_name)
summarytable5 = summarytable5.groupby(["Spending Ranges (Per Student)"]).agg({"Average Math" : "mean", "Average Reading" : "mean", "% Passing Math": "mean", "% Passing Reading": "mean", "% Passing Both": "mean"})
summarytable5
```

## Scores by School Size
