```python
 # Clustering/Dimensionality Reduction
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# DataScience
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read datafram, drop column
df = pd.read_csv("myopia.csv")
target = df.MYOPIC
df.drop(columns=["MYOPIC"], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Check for larger values 
X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("Plot")
plt.show() 
```

```python
#create pca model
pca = PCA()
pca.fit(df)

plt.plot(range(len(df.columns)), pca.explained_variance_ratio_)
plt.xlabel("principal components")
plt.ylabel("explained variance")
plt.title("PCA")
plt.show()
```

```python
df.corr()
```

```python
pca = PCA(n_components=3)
pca.fit(df)
```

```python
#create components of transformation
comps = pd.DataFrame(pca.transform(df), columns=["pca1", "pca2", "pca3"])
comps.head()
```

```python
#create tsne plot
X = TSNE(learning_rate="auto").fit_transform(comps)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("tsne plot")
plt.show()
```

```python
#elbow plot of inertia with dataframe
inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("clusters")
plt.ylabel("inertia")
plt.title("elbow plot df")

plt.show()
```

```python
#elbow plot of inertia with component dataframe
inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(comps)
    
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("clusters")
plt.ylabel("inertia")
plt.title("elbow plot comps")

plt.show()
```

```python
#tsne plot with dataframe 8 clusters
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("tsne plot df")
plt.show()
```

```python
#tsne plot with dataframe 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("tsne plot df")
plt.show()
```

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    affinity="euclidean",
    linkage="ward"
)

preds = cluster.fit_predict(df)
plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("TSNE Plot")
plt.show()
```
